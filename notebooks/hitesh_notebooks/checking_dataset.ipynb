# Dropping null rows from dataframe 2

```python
# Print the number of rows before dropping
print("Number of airports before dropping null rows:", len(df2))

# Drop rows with missing coordinates
df2 = df2.dropna(subset=['LATITUDE', 'LONGITUDE'])

# Print the number of rows after dropping
print("Number of airports after dropping null rows:", len(df2))

# Verify no more null values in coordinates
print("\nRemaining null values in airports dataset:")
print(df2.isnull().sum())
```

# Merging all 3 dataframes

```python
# First, merge flights with airlines
flights_airlines = df3.merge(df1, left_on='AIRLINE', right_on='IATA_CODE', how='left')
print("Shape after merging with airlines:", flights_airlines.shape)
flights_airlines
```

```python
# Now merge with airports for origin airport information
flights_with_origin = flights_airlines.merge(
    df2[['IATA_CODE', 'AIRPORT', 'CITY', 'STATE', 'COUNTRY', 'LATITUDE', 'LONGITUDE']], 
    left_on='ORIGIN_AIRPORT', 
    right_on='IATA_CODE', 
    how='left',
    suffixes=('', '_origin')
)
print("\nShape after merging with origin airports:", flights_with_origin.shape)
flights_with_origin
```

```python
# Finally merge with airports for destination airport information
flights_full = flights_with_origin.merge(
    df2[['IATA_CODE', 'AIRPORT', 'CITY', 'STATE', 'COUNTRY', 'LATITUDE', 'LONGITUDE']], 
    left_on='DESTINATION_AIRPORT', 
    right_on='IATA_CODE', 
    how='left',
    suffixes=('', '_dest')
)
print("\nFinal shape after all merges:", flights_full.shape)
flights_full
```

```python
# Display the columns in our final dataset
print("\nColumns in the final merged dataset:")
print(flights_full.columns.tolist())

# Check for any missing values
print("\nMissing values in the merged dataset:")
print(flights_full.isnull().sum())
```
