```python
import polars as pl
import numpy as np
```

```python
# Show all rows
pl.Config.set_tbl_rows(-1)

# Show all columns
pl.Config.set_tbl_cols(-1)

# Widen the column display so you can see more of the content
pl.Config.set_tbl_width_chars(500)

# Set the max number of characters to display for string columns
pl.Config.set_fmt_str_lengths(100)
```

```python
lf = pl.scan_parquet("data/flights_preprocessed_no_delay.parquet")
lf.head().collect()
```

```python
value_counts_dict = {}
for col in list(lf.collect_schema()):
    value_counts_dict[col] = lf.select(pl.col(col).unique()).collect()
value_counts_dict
```
